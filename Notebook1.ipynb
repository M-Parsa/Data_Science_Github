```python
import pandas as pd

df = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie'],
    'sales_2023': [250, 300, 150],
    'sales_2024': [270, 320, 180],

})

print(df)
```

```python
df_long = pd.melt(df, id_vars=['id', 'name'], var_name='year', value_name='sales')
df_long
```

```python
pd.pivot_table(df_long, index=['id', 'name'], columns='year', values='sales')
```

```python
import pandas as pd

df = pd.DataFrame({
    'id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie'],
    'sales_2023': [250, 300, 150],
    'sales_2024': [270, 320, 180],
    'revenue_2023': [5000, 7000, 4000],
    'revenue_2024': [5400, 7500, 4500]
})
df
```

```python
df_long = pd.wide_to_long(df,
                          stubnames=['sales', 'revenue'],
                          i=['id', 'name'],
                          j='year',
                          sep='_').reset_index()
print(df_long)
```

```python
# Reshape long to wide

df_wide = df_long.pivot_table(index=['id', 'name'], columns=['year'], values=['sales', 'revenue']).reset_index()
df_wide
```
